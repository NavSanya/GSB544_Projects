# Description
## Data
This dataset consists of information about properties sold in Ames, Iowa. You may have seen this data before, and you may find examples of it being used elsewhere online. The version in this competition has many of the variables removed, for uniqueness and simplicity.

## Data fields
Here's a brief description of the variables:

SalePrice - the property's sale price in dollars. This is the target variable that you're trying to predict.
PID - An identifying number of the property
LotFrontage: Linear feet of street connected to property
LotArea: Lot size in square feet
Street: Type of road access
Neighborhood: Physical locations within Ames city limits
BldgType: Type of dwelling
HouseStyle: Style of dwelling
OverallQual: Overall material and finish quality
OverallCond: Overall condition rating
YearBuilt: Original construction date
RoofStyle: Type of roof
Heating: Type of heating
CentralAir: Central air conditioning
Electrical: Electrical system
GrLivArea: Above grade (ground) living area square feet
FullBath: Full bathrooms above grade
HalfBath: Half baths above grade
Bedroom AbvGr: Number of bedrooms above basement level
TotRmsAbvGrd: Total rooms above grade (does not include bathrooms)
Functional: Home functionality rating
ScreenPorch: Screen porch area in square feet
PoolArea: Pool area in square feet
YrSold: Year Sold
SaleType: Type of sale

## GOAL
Your mission today is to use the available variables to predict the sales price of homes.

Only your first two submissions will be graded!
Acknowledgments
The Ames Housing dataset was compiled by Dean De Cock for use in data science education. It's an incredible alternative for data scientists looking for a modernized and expanded version of the often cited Boston Housing dataset.

## Evaluation
### Goal
It is your job to predict the sales price for each house. For each PID in the test set, you must predict the value of the SalePrice variable.

Submission File Format
The file should contain a header and have the following format:

PID             SalePrice
907135180	169000.1
903206120	187724.1233
etc.

You can download an example submission file (sample_submission.csv) on the Data page.

### Metric
Submissions are evaluated on Root-Mean-Squared-Error (RMSE) between the logarithm of the predicted value and the logarithm of the observed sales price. (Taking logs means that errors in predicting expensive houses and cheap houses will affect the result equally.)

Note that although this metric is calculated on log price, your submitted predictions need to be regular prices.

## Grading
Your grade on this portion of the Final will be determined by the metric achieved by your submitted predictions.
