```python
def plot_species_island(species = None, island=None):
    """
    Dynamic lookup. You should know cloumn Names.
    """

    new_penguins = penguins

    if species is not None:
        # Filter for species
        new_penguins = new_penguins[new_penguins['species'] == species]

    # If island is provided, filter for that island too
    if island is not None:
        new_penguins = new_penguins[new_penguins["island"] == island]

    plot = (
        p9.ggplot(new_penguins, p9.aes(x='bill_length_mm', y="bill_depth_mm", fill = "island"))
        + p9.geom_point()
        + p9.labs(
            title="Scatter Plot",
            x="bill_length", y="bill_depth"
        )
    )
    plot.show()
```

```python
plot_species_island(species="Adelie")
```

```python
def rescale_vector(vec):

  if any(np.isnan(vec)):
        exit("What the hell!")

  min_vec = min(vec)
  print("min: " + str(min_vec))

  max_vec = max(vec)
  print("max: " + str(max_vec))

  rescaled = (vec - min_vec)/(max_vec - min_vec)

  return rescaled
```

```python
my_list = [-1, 2, 3, 4]
new_list = np.sqrt(my_list)
# print(np.isnan(new_list))
print(rescale_vector(new_list))
```

```python
def times_seven(x):

  if not isinstance(x, (int, float)):
    exit("Error: Input must be numeric")

  if x == 7:
    print("I love sevens!")

  return x * 7
```
